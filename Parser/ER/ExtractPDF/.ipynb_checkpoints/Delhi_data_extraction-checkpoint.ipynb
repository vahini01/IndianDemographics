```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Sat Jan  6 12:53:49 2018

@author: dhingratul
"""
import time
from bs4 import BeautifulSoup
import urllib
import ssl
import requests
import wget
import os

def download_file_W_delhi(pdf_url, mdir, filename, flag=False):
    filename = mdir + filename
    ssl._create_default_https_context = ssl._create_unverified_context
    wget.download(pdf_url, filename)
    if os.stat(filename).st_size == 0:
        flag = 0
    else:
        flag = 1
    return flag


mdir = '../Delhi/'
base_url = "https://ceodelhi.gov.in/engdata/"
location = ""
files = os.listdir(mdir)
files_list = []
for file in files:
  if ".pdf" in file:
    files_list.append(file)

i_start = 51
j_start = 1
arr = [296, 305, 184, 162, 217, 254, 344, 279, 274, 173, 248, 180, 171, 164, 143, 154, 182, 164, 174, 133, 131, 152,179, 189]
arr1 = [169, 163, 166, 165, 144, 198, 376, 259, 191, 375, 257, 192, 239, 115, 177, 166, 129, 148, 145, 155, 180, 176, 210, 140, 151, 
       163, 161, 161, 262, 265, 161, 178, 197, 179, 184, 203, 177, 184, 178, 185, 168, 200, 204, 218, 254, 269]
for i in range(i_start, 71):
    if i==51:
        j = 127
    else:
        j = 1
    for j in range(j_start, arr1[i-25]+1):
        print("\n", i, j)
        suffix = "AC{}/U05A{}P{}.pdf".format(i, i, j)
        url = base_url + suffix
#         print(url)
        fid = suffix.replace("/", "_")
        try:
            if fid not in files_list:
                flag = download_file_W_delhi(url, mdir, fid)
                if flag == 0:
                    with open("Delhi.txt", "a") as myfile:
                        myfile.write(url + '\n')
        except urllib.error.HTTPError:
            with open("Delhi.txt", "a") as myfile:
                myfile.write(url + '\n')
```
