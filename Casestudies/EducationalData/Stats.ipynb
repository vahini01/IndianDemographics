```python
#  Unique names in CBSE
import pandas as pd

data = []
for aieee in AL:
    df = pd.read_csv(aieee)
    names = pd.DataFrame()
    names['name'] = df['name']
    size = [aieee[-8:-4], names.shape[0], names.drop_duplicates().shape[0]]
    data.append(size)
    
aieee_data = pd.DataFrame(data, columns=['AIEEE year', 'Total Names', 'Unique Names' ])

all_years = ['ALL', aieee_data['Total Names'].sum(), aieee_data['Unique Names'].sum()]
data.append(all_years)
aieee_data = pd.DataFrame(data, columns=['AIEEE year', 'Total Names', 'Unique Names' ])

aieee_data.head(20)
```
