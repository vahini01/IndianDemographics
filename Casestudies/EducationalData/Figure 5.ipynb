```python
import pandas as pd

reserved_ratio = []
women_ratio = []

for i in range(7):
    c12 = pd.read_csv(C12LM[2+i])
    c10 = pd.read_csv(C10LM[i])
    r12  = c12.groupby('Caste').count()['name'][1] / (c12.groupby('Caste').count()['name'][1] + c12.groupby('Caste').count()['name'][0] )
    r10 = c10.groupby('Caste').count()['name'][1] / (c10.groupby('Caste').count()['name'][1] + c10.groupby('Caste').count()['name'][0] )
    reserved_ratio.append(round(round(r12*100,2)/round(r10*100, 2) , 2))
    w12  = c12.groupby('Gender').count()['name'][1] / (c12.groupby('Gender').count()['name'][1] + c12.groupby('Gender').count()['name'][0] )
    w10 = c10.groupby('Gender').count()['name'][1] / (c10.groupby('Gender').count()['name'][1] + c10.groupby('Gender').count()['name'][0] )
    women_ratio.append(round(round(w12*100,2)/round(w10*100, 2) , 2))
    
```

```python
print(reserved_ratio)
print(women_ratio)
```
